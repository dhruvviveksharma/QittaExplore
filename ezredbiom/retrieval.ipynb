```python
import os
os.environ['QIITA_CONFIG_FP'] = '/home/d4sharma/qiita-web/qiita_config.cfg'

from qiita_db.sql_connection import TRN
import pandas as pd
with TRN:
    # Get sample IDs for a study
    TRN.add("SELECT sample_id FROM qiita.study_sample WHERE study_id = 77")
    sample_ids = TRN.execute_fetchindex()
print("Sample IDs:", sample_ids[:10])

# Look inside sample 77.TS1 (study 77's sample table)
with TRN:
    TRN.add("SELECT * FROM qiita.sample_77 WHERE sample_id = '77.TS1'")
    sample = TRN.execute_fetchindex()
print("Sample 77.TS1:", sample)
```

```python
with TRN:
    TRN.add("SELECT * FROM qiita.sample_77 WHERE sample_id = '77.TS1'")
    rows = TRN.execute_fetchindex()

# Column names
with TRN:
    TRN.add("""
        SELECT column_name FROM information_schema.columns
        WHERE table_schema = 'qiita' AND table_name = 'sample_77'
        ORDER BY ordinal_position
    """)
    cols = [k[0] for k,v in TRN.execute_fetchindex()]

df_sample = pd.DataFrame(rows, columns=cols)
print(df_sample)
```

```python
with TRN:
    TRN.add("""
        SELECT table_name 
        FROM information_schema.tables 
        WHERE table_schema = 'qiita'
        ORDER BY table_name
    """)
    tables = TRN.execute_fetchindex()

tables
```

```python
with open('qiita_tables.txt', 'w') as f:
    for table in tables:
        f.write(table[0] + '\n')
```

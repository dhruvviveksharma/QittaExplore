```python
import os
os.environ['QIITA_CONFIG_FP'] = '/home/d4sharma/qiita-web/qiita_config.cfg'

from qiita_db.sql_connection import TRN
import pandas as pd
with TRN:
    # Get sample IDs for a study
    TRN.add("SELECT sample_id FROM qiita.study_sample WHERE study_id = 77")
    sample_ids = TRN.execute_fetchindex()
print("Sample IDs:", sample_ids[:10])

# Look inside sample 77.TS1 (study 77's sample table)
with TRN:
    TRN.add("SELECT * FROM qiita.sample_77 WHERE sample_id = '77.TS1'")
    sample = TRN.execute_fetchindex()
print("Sample 77.TS1:", sample)

# Artifact 279
with TRN:
    TRN.add("""
        SELECT a.artifact_id, at.artifact_type, a.visibility_id, a.command_id, a.generated_timestamp
        FROM qiita.artifact a
        JOIN qiita.artifact_type at USING (artifact_type_id)
        WHERE a.artifact_id = 279
    """)
    artifact = TRN.execute_fetchindex()
print("Artifact 279:", artifact)
```
